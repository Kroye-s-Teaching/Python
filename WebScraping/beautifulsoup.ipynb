# Beautiful Soup
A Beautiful Soup egy konyvtarcsomag html kod feldolgozasara, ami egyszerubbe teszi a beolvasott html kezeleset.

https://www.digitalocean.com/community/tutorials/how-to-work-with-web-data-using-requests-and-beautiful-soup-with-python-3

```python
# syntax: bs4 modulbol betoltjuk a BeautifulSoup csomagot
from bs4 import BeautifulSoup
import csv
import pandas as pd
import requests
```

```python
url='https://oilprice.com/'
```

```python
page = requests.get(url)
```

```python
# BeautfilSoup: input - feldolgozando str, parser amivel feldolgozzuk
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup
```

```python
# prettify: kiiratast javito fuggveny
print(soup.prettify())
```

```python
# a find_all fuggveny egy listaban minden elofordulast megtalal egy adott html tag-re
soup.find_all('td')[2]
```

```python
# a get_text-el hivatkozhatunk a tag-eken beluli ertekre
soup.find_all('td')[2].get_text()
```

```python
# Lekerdezzuk a kulonbozo reszvenyek neveit
blends = soup.find_all(class_='blend_name')
```

```python
blends
```

```python
# Majd lekerdezzuk az ezekhez tartozo arakat
vals = soup.find_all(class_='value')
```

```python
# Itt vezetjuk be a zip-et
for blend,val in zip(blends,vals):
    print(blend.contents[0],val.contents[0])
```

```python
# A zip osszefuzi, osszekapcsolja elemenkent a listakat es egy uj listat keszit, amiben az elemek a listak 0., 1., ..
# elemeibol alkotott tuple-ok (megvaltoztathatatlan listak)
z = zip(['a','b','c'],[1,2,3,4],[10,20,30])
```

```python
list(z)
```

```python
# L lista vs T tuple
L = [1,2,3]
L[1] = 20
print(L)
```

A csv modul hasznalhato fajl irasra, olvasasra.

Irjuk ki a beolvasott ertekparokat egy csv file-ba!

- open - fajl neve (eleresi uttal), megnyitas tipusa (r - read, w- write, a- append)
- csv.writer - iro objektum a csv modulbol, alapertelmezesben ,-vel elvalasztottan ir, de allithato
- writerow - a writer objektum egy fuggvenye, ami egy listat var
- close - fajl bezarasa

```python
csvfile = open("oil_prices.csv", "w+")
mywriter = csv.writer(csvfile)
mywriter.writerow(["Name","Price"])
for blend,val in zip(blends,vals):
    mywriter.writerow([blend.contents[0],val.contents[0]])
csvfile.close()
```

Ugyanez kivetelkezelessel, try-finally-t hasznalva. Hiba eseten az eredeti hibat fogja dobni.

```python
csvfile = open("oil_prices.csv", "w")
try:
    mywriter = csv.writer(csvfile)
    mywriter.writerow(["Name","Price"])
    for blend,val in zip(blends,vals):
        mywriter.writerow([blend.contents[0],val.contents[0]])
finally:
    csvfile.close()
```

A with egy context manager, ami lenyegeben a fenti megoldas egy roviditese.

Minden osztalynak es leszarmazott objektumnak van egy __enter__ es egy __exit__ metodusa.

A with ezeket kapcsolja ossze, kezdetben lefuttatja az belepesi metodust, jelen esetben megnyitja a fajlt es a with block vegen lezarja, ez a fajlmuvelethez rendelt exit, ami lefut a finally-ben. Tehat biztosak lehetunk benne, hogy nem hagytuk veletlenul megnyitva az irt fajlt.

https://book.pythontips.com/en/latest/context_managers.html

```python
with open("oil_prices.csv", "w") as csvfile:
    mywriter = csv.writer(csvfile)
    mywriter.writerow(["Name","Price"])
    for blend,val in zip(blends,vals):
        mywriter.writerow([blend.contents[0],val.contents[0]])
```
